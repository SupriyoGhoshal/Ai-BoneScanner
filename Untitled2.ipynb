```python
from PIL import Image
import os

dataset_path = '/kaggle/input/multi-class-knee-osteoporosis-x-ray-dataset/OS Collected Data'

# Example: open first image in Normal
normal_folder = os.path.join(dataset_path, 'Normal')
first_image = os.listdir(normal_folder)[0]
image_path = os.path.join(normal_folder, first_image)

img = Image.open(image_path)
img.show()
```

```python
import numpy as np
from PIL import Image

categories = ['Normal', 'Osteopenia', 'Osteoporosis']
data = []
labels = []

for category in categories:
    folder_path = os.path.join(dataset_path, category)
    for file in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file)
        try:
            img = Image.open(file_path).convert('RGB').resize((224,224))
            data.append(np.array(img))
            labels.append(category)
        except:
            print("Error loading", file_path)

data = np.array(data)
labels = np.array(labels)
print("Data shape:", data.shape)
print("Labels shape:", labels.shape)
```

```python
dataset_path = "/kaggle/input/multi-class-knee-osteoporosis-x-ray-dataset/OS Collected Data"
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os

train_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2,
    rotation_range=15,
    zoom_range=0.1,
    horizontal_flip=True
)

# --- Start of diagnostic additions ---
print(f"Verifying dataset_path: {dataset_path}")
if not os.path.exists(dataset_path):
    print(f"Error: The directory '{dataset_path}' does not exist.")
else:
    print(f"Success: The directory '{dataset_path}' exists.")
    try:
        # Attempt to list directory contents to further diagnose access issues
        contents = os.listdir(dataset_path)
        print(f"Contents of '{dataset_path}': First 5 items: {contents[:5]}")
        if not any(os.path.isdir(os.path.join(dataset_path, item)) for item in contents):
            print("Warning: No subdirectories found in dataset_path. `flow_from_directory` expects class subfolders.")
    except FileNotFoundError as e:
        print(f"Error: Still unable to list contents of '{dataset_path}'. Original error: {e}")
    except Exception as e:
        print(f"An unexpected error occurred while trying to list directory contents: {e}")
# --- End of diagnostic additions ---

train_gen = train_datagen.flow_from_directory(
    dataset_path,
    target_size=(224,224),
    batch_size=32,
    class_mode='categorical',
    subset='training'
)

val_gen = train_datagen.flow_from_directory(
    dataset_path,
    target_size=(224,224),
    batch_size=32,
    class_mode='categorical',
    subset='validation'
)
```

```python
#model traning
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, Dropout
from tensorflow.keras.models import Model
import matplotlib.pyplot as plt
```

```python
base = EfficientNetB0(
    weights='imagenet',
    include_top=False,
    input_shape=(224,224,3)
)

base.trainable = False   # freeze feature extractor
```

```python
x = GlobalAveragePooling2D()(base.output)
x = Dropout(0.3)(x)
x = Dense(256, activation='relu')(x)
x = Dropout(0.3)(x)
output = Dense(train_gen.num_classes, activation='softmax')(x)

model = Model(inputs=base.input, outputs=output)

model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()
```

```python
history = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=10
)
```

```python
model.save("nutribone_effnet_model.h5")
print("Model Saved!")
```

```python
#Stage 2,
#Evaluate model (accuracy, loss, plots, confusion matrix, classification report)

import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

# history from model.fit
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(history.history['loss'], label='train_loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.legend(); plt.title('Loss')

plt.subplot(1,2,2)
plt.plot(history.history['accuracy'], label='train_acc')
plt.plot(history.history['val_accuracy'], label='val_acc')
plt.legend(); plt.title('Accuracy')
plt.show()

# Get true labels and preds on validation set
val_gen.reset()
preds = model.predict(val_gen, verbose=1)
y_pred = np.argmax(preds, axis=1)
y_true = val_gen.classes
class_names = list(val_gen.class_indices.keys())

print(classification_report(y_true, y_pred, target_names=class_names))

cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicted'); plt.ylabel('True'); plt.title('Confusion Matrix')
plt.show()
```
